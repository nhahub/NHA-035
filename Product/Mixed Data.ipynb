```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
 
 
import numpy as np
import time
import random
import pandas as pd
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
driver.get(f"https://www.amazon.eg/s?i=electronics&rh=n%3A21832887031&s=popularity-rank&fs=true&language=en&ref=lp_21832887031_sar") 
```

```python
data = []


for page_num in range(8, 11):
    time.sleep(random.uniform(3,7))
    driver.get(f"https://www.amazon.eg/-/en/s?i=electronics&rh=n%3A21832887031&s=popularity-rank&fs=true&page={page_num}&language=en&qid=1759743428&xpid=ZsUd_BzzbA__g&ref=sr_pg_{page_num}")
    time.sleep(random.uniform(2, 5))
    
    products = driver.find_elements(By.CSS_SELECTOR, "div.s-main-slot div[data-component-type='s-search-result']")

    for product in products:
        
        try:
            sponsored = product.find_element(By.XPATH, ".//*[contains(text(), 'Sponsored')]")
            print("Skipping sponsored product")
            continue
        except:
            pass
        
        title = None
        link = None
        price = np.nan
        rating = "No rating"
        num_of_reviews = "0"
        asin = None
        
        try:
            title = product.find_element(By.CSS_SELECTOR, "a h2 span").text.strip()
        except:
            title = None
        
        try:
            link = product.find_element(By.CSS_SELECTOR, ".a-link-normal.s-line-clamp-4.s-link-style.a-text-normal").get_attribute('href')
        except:
            link = None
        
        if link:   
            try:
                asin = link.split("/dp/")[1].split("/")[0]
            except:
                asin = None
        
        try:
            price = product.find_element(By.CSS_SELECTOR, ".a-price-whole").text
        except:
            price = np.nan
            
        try:
            rating = product.find_element(By.CSS_SELECTOR, ".a-size-small.a-color-base").get_attribute("textContent")    
        except:
            rating = "No Rating"
        
        try:
            num_of_reviews = product.find_element(By.CSS_SELECTOR, ".a-size-mini.puis-normal-weight-text.s-underline-text").get_attribute("textContent").strip("()")
        except:
            num_of_reviews = "0"

        data.append({
            "Title": title,
            "Link": link,
            "ASIN": asin,
            "Price": price,
            "Rating": rating,
            "Num_Reviews": num_of_reviews,
            "Website": "Amazon"
        })




```

```python
seen_review_count = []
for page_num in range(1, 5):
    time.sleep(random.uniform(2, 5))
    driver.get(f"https://www.noon.com/egypt-en/eg-all-audio/?f%5BisCarousel%5D=true&f%5Bis_fbn%5D%5B%5D=1&sort%5Bby%5D=best_rating&sort%5Bdir%5D=desc&limit=50&page={page_num}&isCarouselView=false")
    time.sleep(random.uniform(2, 5))
    
    products = driver.find_elements(By.CLASS_NAME, "PBoxLinkHandler_linkWrapper__8FlHB")

    for product in products:
        
        product_title = None
        product_link = None
        product_price = np.nan
        product_rating= "No rating"
        review_count = "0"
        asin = None
        
        try:
            review_count = product.find_element(By.CSS_SELECTOR, "div.RatingPreviewStar_countCtr__p_2z7 span").text
            if review_count in seen_review_count:
                print("Skipping duplicated reviews...")
                continue
            else:
                seen_review_count.append(review_count)
        except:
            review_count = "0"
        
        try:
            product_title = product.find_element(By.CLASS_NAME, "ProductDetailsSection_title__JorAV").text
        except:
            product_title = None
            
        try:
            product_link = product.find_element(By.CLASS_NAME, "PBoxLinkHandler_productBoxLink__fAJHN").get_attribute("href")
        except:
            product_link = None
            
        try: 
            product_price = product.find_element(By.CLASS_NAME, "Price_amount__2sXa7").get_attribute("textContent")
        except:
            product_price = np.nan
            
        try:
            product_rating = product.find_element(By.CLASS_NAME, "RatingPreviewStar_textCtr__sfsJG").get_attribute("textContent")
        except:
            product_rating = "No rating"
            
        if product_link:   
            try:
                asin = product_link.split("/")[5]
            except:
                asin = None
        
        time.sleep(random.uniform(0.3, 1))
        
        data.append({
            "Title": product_title,
            "Link": product_link,
            "ASIN": asin,
            "Price": product_price,
            "Rating": product_rating,
            "Num_Reviews": review_count,
            "Website": "Noon"
        })

        
```

```python
star_rating = []
reviews_data = []
translated = False



for item in data:
    if item['Website'] == "Noon":
        reviewed_asin = item['ASIN']
        reviewed_product_title = item['Title']
        if not reviewed_product_title:
            reviewed_product_title = None
        if not reviewed_asin:
            continue
        
        review_link = f"https://www.noon.com/egypt-ar/reviews/{reviewed_asin}"
        driver.get(review_link)
        time.sleep(7)
        
        try:
            if not translated:
                translate_button = driver.find_element(By.CLASS_NAME, "TranslateAllReviewsBtn_translateAllCtr__y_MFV")
                translate_button.click()
                time.sleep(1)
                translated = True
        except:
            print("No translate button found.")
        
        page_num = 1
        
        while True:
            print(f"Scraping {reviewed_asin}, Page: {page_num}")
            
            time.sleep(7)
            
            review_blocks = driver.find_elements(By.CLASS_NAME, "ReviewListSection_inView__vMMSX")
            
            for block in review_blocks:
                stars = 0
                try:
                    rating_stars_unfiltered = block.find_elements(By.CSS_SELECTOR, "button.RatingStars_imgCtr__Ihaqs img")
                    for star in rating_stars_unfiltered:
                        color = star.get_attribute("color")
                        if color != "grey3":
                            stars += 1
                        else: 
                            pass
                    num_of_stars = str(stars)
                except:
                    print("No stars found, skipping...")
                    num_of_stars = "No value"
            
                try:
                    review_text = block.find_element(By.CLASS_NAME, "CollapsibleText_textContainer__jUZvK").get_attribute("textContent")
                except:
                    review_text = "No Review"
                    
                if review_text != "No Review":
                    reviews_data.append({
                        "Title": reviewed_product_title,
                        "ASIN": reviewed_asin,
                        "Stars": num_of_stars,
                        "Review": review_text,
                        "Website": "Noon"
                    })
                
            try:
                next_button = driver.find_element(By.CSS_SELECTOR, "li.next")
            
                if "PaginationNumbered_disabled__gO8m6" in  next_button.get_attribute("class"):
                    print(f"No more pages, scraped {page_num} pages from {reviewed_asin}.")
                    break
                
                driver.execute_script("arguments[0].scrollIntoView(true);", next_button)
                time.sleep(2)
                next_button.click()
                
                page_num += 1
                
                if page_num >= 10:
                    break
            except:
                print(f"No next button found for {reviewed_asin}, resuming.")
                break
    else:
        pass
```

```python
reviews_data = []

for item in data:
    if item['Website'] == "Amazon":
        reviewed_asin = item['ASIN']
        reviewed_product_title = item['Title']
        if not reviewed_product_title:
            reviewed_product_title = None
        if not reviewed_asin:
            continue
        
        review_link = f"https://www.amazon.eg/product-reviews/{reviewed_asin}"
        driver.get(review_link)
        time.sleep(random.uniform(3, 7))
        
        page_num = 1
        
        while True:
            print(f"Scraping {reviewed_asin}, Page: {page_num}")
            
            time.sleep(7)
        
            review_blocks = driver.find_elements(By.CSS_SELECTOR, "li[data-hook='review']")
            
            for block in review_blocks:
                try:
                    review_text = block.find_element(By.CSS_SELECTOR, "[data-hook='review-body'] .cr-original-review-content").text
                except:
                    review_text = "No Review"
                
                try:
                    num_of_stars = block.find_element(By.CSS_SELECTOR, ".a-icon-alt").get_attribute("textContent").strip().split()[0]
                except:
                    num_of_stars = "No value"
            

            
                if review_text != "No Review":
                    reviews_data.append({
                        "Title": reviewed_product_title,
                        "ASIN": reviewed_asin,
                        "Stars": num_of_stars,
                        "Review": review_text,
                        "Website": "Amazon"
                    })
                    
            try:
                next_button = driver.find_element(By.CSS_SELECTOR, "li.a-last a")
                next_button.click()
                page_num += 1
                time.sleep(random.uniform(2,5))
            except:
                print("Button is disabled, this is the last page.")
                break
            
            if page_num>12:
                break
    else:
        pass
```
