```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
 
 
import numpy as np
import time
import random
import pandas as pd
```

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
```

```python
driver.get(f"https://www.amazon.eg/s?i=electronics&rh=n%3A21832887031&s=popularity-rank&fs=true&language=en&ref=lp_21832887031_sar") 
```

```python
data = []


for page_num in range(1, 5):
    time.sleep(random.uniform(3,7))
    driver.get(f"https://www.amazon.eg/-/en/s?i=electronics&rh=n%3A21832887031&s=popularity-rank&fs=true&page={page_num}&language=en&qid=1759743428&xpid=ZsUd_BzzbA__g&ref=sr_pg_{page_num}")
    time.sleep(random.uniform(2, 5))
    
    products = driver.find_elements(By.CSS_SELECTOR, "div.s-main-slot div[data-component-type='s-search-result']")

    for product in products:
        
        try:
            sponsored = product.find_element(By.XPATH, ".//*[contains(text(), 'Sponsored')]")
            print("Skipping sponsored product")
            continue
        except:
            pass
        
        title = None
        link = None
        price = np.nan
        rating = "No rating"
        num_of_reviews = "0"
        asin = None
        
        try:
            title = product.find_element(By.CSS_SELECTOR, "a h2 span").text.strip()
        except:
            title = None
        
        try:
            link = product.find_element(By.CSS_SELECTOR, ".a-link-normal.s-line-clamp-4.s-link-style.a-text-normal").get_attribute('href')
        except:
            link = None
        
        if link:   
            try:
                asin = link.split("/dp/")[1].split("/")[0]
            except:
                asin = None
        
        try:
            price = product.find_element(By.CSS_SELECTOR, ".a-price-whole").text
        except:
            price = np.nan
            
        try:
            rating = product.find_element(By.CSS_SELECTOR, ".a-size-small.a-color-base").get_attribute("textContent")    
        except:
            rating = "No Rating"
        
        try:
            num_of_reviews = product.find_element(By.CSS_SELECTOR, ".a-size-mini.puis-normal-weight-text.s-underline-text").get_attribute("textContent").strip("()")
        except:
            num_of_reviews = "0"

        data.append({
            "Title": title,
            "Link": link,
            "ASIN": asin,
            "Price": price,
            "Rating": rating,
            "Num_Reviews": num_of_reviews
        })




```

```python
star_rating = []
reviews_data = []

for item in data:
    reviewed_asin = item['ASIN']
    reviewed_product_title = item['Title']
    if not reviewed_product_title:
        reviewed_product_title = None
    if not reviewed_asin:
        continue
    
    review_link = f"https://www.amazon.eg/product-reviews/{reviewed_asin}"
    driver.get(review_link)
    time.sleep(random.uniform(3, 7))
    
    page_num = 1
    
    while True:
        print(f"Scraping {reviewed_asin}, Page: {page_num}")
        
        time.sleep(7)
    
        review_blocks = driver.find_elements(By.CSS_SELECTOR, "li[data-hook='review']")
        
        for block in review_blocks:
            try:
                review_text = block.find_element(By.CSS_SELECTOR, "[data-hook='review-body'] .cr-original-review-content").text
            except:
                review_text = "No Review"
            
            try:
                num_of_stars = block.find_element(By.CSS_SELECTOR, ".a-icon-alt").get_attribute("textContent").strip().split()[0]
            except:
                num_of_stars = "No value"
        

        
            if review_text != "No Review":
                reviews_data.append({
                    "Title": reviewed_product_title,
                    "ASIN": reviewed_asin,
                    "Stars": num_of_stars,
                    "Review": review_text
                })
                
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, "li.a-last a")
            next_button.click()
            page_num += 1
            time.sleep(random.uniform(2,5))
        except:
            print("Button is disabled, this is the last page.")
            break
        
        if page_num>10:
            break
        
    
```
